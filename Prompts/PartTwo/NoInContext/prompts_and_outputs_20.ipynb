````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Data: Percentage of organizations by country
country_counts = combined_df['Country'].value_counts(normalize=True) * 100

# Create the plot
plt.figure(figsize=(12, 8))  # Increase the figure size for better clarity
ax = country_counts.plot(kind='bar', color='skyblue', alpha=0.85)

# Set titles and labels with increased font sizes for better readability
plt.title('Percentage of Organizations by Country', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Percentage of Organizations (%)', fontsize=14)

# Rotate x-tick labels to prevent overlap and ensure they are readable
plt.xticks(rotation=45, fontsize=12, ha="right", rotation_mode="anchor")

# Draw a grid for the y-axis to improve readability of the y-values
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Remove the top and right spines to make the plot look cleaner
sns.despine(top=True, right=True)

# Annotate bars with their respective values to provide clear information
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 10), textcoords='offset points')

# Show the plot
plt.show()
````
